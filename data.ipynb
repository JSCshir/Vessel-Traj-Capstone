```python
import pandas as pd
import numpy as np
import folium as fl
```

```python
file_path = "obc_tanker_cargo.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
print(df.dtypes)
```

```python
df = df[['MMSI', 'LAT_AVG', 'LON_AVG', 'PERIOD', 'SPEED_KNOTS', 'COG_DEG', 'HEADING_DEG']]
df = df.rename(columns = {'LAT_AVG':'LAT', 'LON_AVG':'LON', 'SPEED_KNOTS':'SPEED', 'COG_DEG':'COG', 'HEADING_DEG':'HEADING'})

print(df.dtypes)
```

```python
df = df.dropna()
print(len(df))
```

```python
MMSI_ids = df['MMSI'].unique()
```

```python
print(MMSI_ids[0])
```

```python
one_vessel_df = df[(df['MMSI']==MMSI_ids[0])]
one_vessel_df.head()
```

```python
one_vessel_df.to_csv('output.txt', sep=' ', index=False)
```
